```python
def two_sum(nums: list[int], target: int) -> list[int]:
    # Stores visited numbers as keys and their indices as values
    seen = {}
    
    for index, num in enumerate(nums):
        complement = target - num
        
        # Check if the needed partner number was already seen
        if complement in seen:
            return [seen[complement], index]
        
        # Track the current number and its index
        seen[num] = index
        
    return []  # Return empty list if no pair is found
```

```python
list = [ 2, 7, 11, 15]
target = 9
result = two_sum(list, target)
```
